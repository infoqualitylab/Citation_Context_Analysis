```python
#EXTRACT NOUNS/NOUN PHRASES/ADJECTIVES from the citation context 
import pandas as pd
import csv
from textblob import TextBlob
from nltk.tokenize import word_tokenize
import nltk

df = pd.read_csv('/Users/phuong/Documents/Citation_Context_Analysis/Analysis/Example_citation_context_RR.csv', encoding = "ISO-8859-1", error_bad_lines=False)
count = 1
citation_context_list = df["Citation Context"]
is_noun = lambda pos: pos[:2] == 'NN'
is_adjective = lambda pos: pos[:2] == 'JJ'

for citation_context in citation_context_list:
    print (count)
    print (citation_context)
    tokenized = (word_tokenize(citation_context))
    nouns = [word for (word, pos) in nltk.pos_tag(tokenized) if is_noun(pos)] 
    adjectives = [word for (word, pos) in nltk.pos_tag(tokenized) if is_adjective(pos)] 
    
    print (nouns)
    print (adjectives)
    
    blob = TextBlob(citation_context)
    print(blob.noun_phrases)
    count +=1
```

```python
#CLUSTER Citation context by TFIDF

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np

documents = citation_context_list

vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(documents)

true_k = 5
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

print("Top terms per cluster:")
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(true_k):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print

 
```
